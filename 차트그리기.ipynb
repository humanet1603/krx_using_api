### 전체기간의 봉차트 그리기

```python
# Install necessary libraries
!pip install pandas plotly
!pip install plotly.express
!pip install plotly.graph_objects

# Import libraries
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px


# Load your data
df = pd.read_csv('/content/1tsm.csv')

# Check the first few rows of the dataframe
print(df.head())

# Create a Figure object
fig = go.Figure()

# Create a candlestick chart
fig.add_trace(go.Candlestick(x=df['date'],
                open=df['open'],
                high=df['high'],
                low=df['low'],
                close=df['close']))


# Updating layout to add titles and labels
fig.update_layout(title='Candlestick Chart for Entire Period',
                  xaxis_title='Date',
                  yaxis_title='Price')

# Show the plot
fig.show()
```
