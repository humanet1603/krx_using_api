```python
import yfinance as yf
import matplotlib.pyplot as plt

# Define the tickers and the time period
tickers = ['NVDA', '^GSPC']
start_date = '2024-01-02'
end_date = '2024-05-22'

# Fetch the data
data = yf.download(tickers, start=start_date, end=end_date)['Close']

# Normalize data to 0 at the start date
data_normalized = (data / data.iloc[0] * 100) - 100

# Calculate moving averages
data_normalized['NVDA_MA5'] = data_normalized['NVDA'].rolling(window=5).mean()
data_normalized['NVDA_MA20'] = data_normalized['NVDA'].rolling(window=20).mean()
data_normalized['^GSPC_MA5'] = data_normalized['^GSPC'].rolling(window=5).mean()
data_normalized['^GSPC_MA20'] = data_normalized['^GSPC'].rolling(window=20).mean()

# Plot the normalized data and moving averages
plt.figure(figsize=(14, 7))

# Plot NVDA
plt.plot(data_normalized.index, data_normalized['NVDA'], label='NVDA')
plt.plot(data_normalized.index, data_normalized['NVDA_MA5'], label='NVDA MA5')
plt.plot(data_normalized.index, data_normalized['NVDA_MA20'], label='NVDA MA20')

# Plot S&P 500
plt.plot(data_normalized.index, data_normalized['^GSPC'], label='S&P 500')
plt.plot(data_normalized.index, data_normalized['^GSPC_MA5'], label='S&P 500 MA5')
plt.plot(data_normalized.index, data_normalized['^GSPC_MA20'], label='S&P 500 MA20')

plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Normalized Value (Starting at 0)')
plt.title('Normalized Stock Prices of NVDA and S&P 500 with MA5 and MA20 (Starting at 0)')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save and display the plot
plt.savefig('nvda_sp500_normalized_zero_ma_plot.png')
plt.show()
```

```python
import yfinance as yf
import pandas as pd

# Define the tickers and the time period
tickers = ['NVDA', '^GSPC']
start_date = '2024-01-02'
end_date = '2024-05-22'

# Fetch the data
data = yf.download(tickers, start=start_date, end=end_date)['Close']

# Normalize data to 0 at the start date
data_normalized = (data / data.iloc[0] * 100) - 100

# Calculate moving averages
data_normalized['NVDA_MA5'] = data_normalized['NVDA'].rolling(window=5).mean()
data_normalized['NVDA_MA20'] = data_normalized['NVDA'].rolling(window=20).mean()
data_normalized['GSPC_MA5'] = data_normalized['^GSPC'].rolling(window=5).mean()
data_normalized['GSPC_MA20'] = data_normalized['^GSPC'].rolling(window=20).mean()

# Find the crossover points for NVDA
nvda_crossover = data_normalized.loc['2024-04-01':].copy()
nvda_crossover['MA5_above_MA20'] = nvda_crossover['NVDA_MA5'] > nvda_crossover['NVDA_MA20']
nvda_crossover['Crossover'] = nvda_crossover['MA5_above_MA20'].diff()

# MA5 crossing above MA20
nvda_golden_cross = nvda_crossover[nvda_crossover['Crossover'] == 1]
# MA5 crossing below MA20
nvda_death_cross = nvda_crossover[nvda_crossover['Crossover'] == -1]

# Find the crossover points for S&P 500
gspc_crossover = data_normalized.loc['2024-04-01':].copy()
gspc_crossover['MA5_above_MA20'] = gspc_crossover['GSPC_MA5'] > gspc_crossover['GSPC_MA20']
gspc_crossover['Crossover'] = gspc_crossover['MA5_above_MA20'].diff()

# MA5 crossing above MA20
gspc_golden_cross = gspc_crossover[gspc_crossover['Crossover'] == 1]
# MA5 crossing below MA20
gspc_death_cross = gspc_crossover[gspc_crossover['Crossover'] == -1]

# Display the results
nvda_golden_cross_dates = nvda_golden_cross[['NVDA', 'NVDA_MA5', 'NVDA_MA20']]
nvda_death_cross_dates = nvda_death_cross[['NVDA', 'NVDA_MA5', 'NVDA_MA20']]
gspc_golden_cross_dates = gspc_golden_cross[['^GSPC', 'GSPC_MA5', 'GSPC_MA20']]
gspc_death_cross_dates = gspc_death_cross[['^GSPC', 'GSPC_MA5', 'GSPC_MA20']]

print("NVDA Golden Cross Dates (MA5 crossing above MA20):")
print(nvda_golden_cross_dates)
print("\nNVDA Death Cross Dates (MA5 crossing below MA20):")
print(nvda_death_cross_dates)

print("\nS&P 500 Golden Cross Dates (MA5 crossing above MA20):")
print(gspc_golden_cross_dates)
print("\nS&P 500 Death Cross Dates (MA5 crossing below MA20):")
print(gspc_death_cross_dates)
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Define the ticker and the time period
ticker = '^GSPC'  # S&P 500
start_date = '2024-01-02'
end_date = '2024-05-24'

# Fetch the data
data = yf.download(ticker, start=start_date, end=end_date)

# Calculate moving averages
data['MA5'] = data['Close'].rolling(window=5).mean()
data['MA20'] = data['Close'].rolling(window=20).mean()

# Filter data for the period from 2024-04-01 to the end date
filtered_data = data[data.index >= '2024-04-01']

# Plot the data
plt.figure(figsize=(14, 7))
plt.plot(filtered_data.index, filtered_data['Close'], label='S&P 500')
plt.plot(filtered_data.index, filtered_data['MA5'], label='S&P 500 MA5')
plt.plot(filtered_data.index, filtered_data['MA20'], label='S&P 500 MA20')

plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('S&P 500 with MA5 and MA20 (2024-04-01 to 2024-05-24)')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save and display the plot
plt.savefig('sp500_ma5_ma20_plot.png')
plt.show()
```
