분단위 1년치 데이터 가져오기

```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

# Ticker symbol
ticker = 'AAPL'

# 시작일 설정 (예: 1년 전)
end_date = datetime.now()
start_date = end_date - timedelta(days=365)

# 60일 간격으로 데이터를 가져와서 결합
all_data = pd.DataFrame()

while start_date < end_date:
    temp_end_date = start_date + timedelta(days=60)
    if temp_end_date > end_date:
        temp_end_date = end_date

    data = yf.download(ticker, start=start_date.strftime('%Y-%m-%d'), end=temp_end_date.strftime('%Y-%m-%d'), interval='1m')
    all_data = pd.concat([all_data, data])

    # 시작일을 업데이트
    start_date = temp_end_date

# 중복된 데이터 제거 (혹시 있을 경우)
all_data = all_data[~all_data.index.duplicated(keep='first')]

print(all_data)
```

```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

# Ticker symbol
ticker = 'AAPL'

# 현재 날짜와 1년 전 날짜 설정
end_date = datetime.now()
start_date = end_date - timedelta(days=365)

# 빈 데이터프레임 생성
all_data = pd.DataFrame()

# 주 단위로 데이터를 가져와서 결합
while start_date < end_date:
    temp_end_date = start_date + timedelta(days=7)
    if temp_end_date > end_date:
        temp_end_date = end_date

    # 데이터 다운로드
    data = yf.download(ticker, start=start_date.strftime('%Y-%m-%d'), end=temp_end_date.strftime('%Y-%m-%d'), interval='1m')

    # 데이터가 비어있지 않으면 처리
    if not data.empty:
        # 거래 시간 필터링 (미국 동부 시간 기준)
        data = data.between_time('09:30', '16:00')
        # 주말 제거
        data = data[data.index.dayofweek < 5]
        # 데이터 결합
        all_data = pd.concat([all_data, data])

    # 시작일을 업데이트
    start_date = temp_end_date

# 중복된 데이터 제거 (혹시 있을 경우)
all_data = all_data[~all_data.index.duplicated(keep='first')]

# 데이터 저장
all_data.to_csv('/content/AAPL.csv')

print("데이터가 /content/AAPL.csv에 저장되었습니다.")
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Base URL of the webpage to scrape
base_url = 'https://stockanalysis.com/markets/gainers/ytd/'

# Function to get the data from a single page
def get_data_from_page(url):
    response = requests.get(url)
    if response.status_code != 200:
        raise Exception(f"Failed to load page {url}")

    soup = BeautifulSoup(response.content, 'html.parser')
    table = soup.find('table')

    # Extract table headers
    headers = [header.text for header in table.find_all('th')]

    # Extract table rows
    rows = []
    for row in table.find_all('tr')[1:]:  # Skip the header row
        cols = row.find_all('td')
        cols = [col.text.strip() for col in cols]
        rows.append(cols)

    return headers, rows

# Initialize an empty DataFrame
all_data = pd.DataFrame()

# Start with the first page
current_page = 1

while True:
    # Construct the URL for the current page
    url = f"{base_url}?page={current_page}"

    try:
        headers, rows = get_data_from_page(url)
    except Exception as e:
        print(e)
        break

    # If no rows are found, we've reached the end
    if not rows:
        break

    # Create a DataFrame from the extracted data
    page_data = pd.DataFrame(rows, columns=headers)

    # Append the data to the main DataFrame
    all_data = pd.concat([all_data, page_data], ignore_index=True)

    print(f"Page {current_page} data collected.")

    # Move to the next page
    current_page += 1

# Save the DataFrame to a CSV file
csv_file_path = '/content/YTD_Gainers.csv'
all_data.to_csv(csv_file_path, index=False)

print(f"Data has been saved to {csv_file_path}")
```
