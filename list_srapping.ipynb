```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Base URL of the webpage to scrape
base_url = 'https://stockanalysis.com/stocks/screener/'

# Function to get the data from a single page
def get_data_from_page(url):
    response = requests.get(url)
    if response.status_code != 200:
        raise Exception(f"Failed to load page {url}")

    soup = BeautifulSoup(response.content, 'html.parser')
    table = soup.find('table', class_='stock-table')  # Find the table with the specific class

    if not table:
        raise Exception(f"No table found on page {url}")

    # Extract table headers
    headers = [header.text.strip() for header in table.find('thead').find_all('th')]

    # Extract table rows
    rows = []
    for row in table.find('tbody').find_all('tr'):  # Iterate over rows in tbody
        cols = row.find_all('td')
        cols = [col.text.strip() for col in cols]
        rows.append(cols)

    return headers, rows

# Initialize an empty DataFrame
all_data = pd.DataFrame()

# Total number of pages
total_pages = 106

# Loop through all pages and collect data
for current_page in range(1, total_pages + 1):
    # Construct the URL for the current page
    url = f"{base_url}?page={current_page}"

    try:
        headers, rows = get_data_from_page(url)
    except Exception as e:
        print(e)
        break

    if not rows:
        print(f"No data found on page {current_page}. Stopping.")
        break

    # Create a DataFrame from the extracted data
    page_data = pd.DataFrame(rows, columns=headers)

    # Append the data to the main DataFrame
    all_data = pd.concat([all_data, page_data], ignore_index=True)

    print(f"Page {current_page} data collected.")

# Remove duplicate rows
all_data.drop_duplicates(inplace=True)

# Save the DataFrame to a CSV file
csv_file_path = '/content/gainers.csv'
all_data.to_csv(csv_file_path, index=False)

print(f"Data has been saved to {csv_file_path}")
```

```python
import os
import pandas as pd

# Directory where CSV files are saved
directory = '/content/'

# Initialize an Excel writer
combined_csv_path = '/content/combined.xlsx'
writer = pd.ExcelWriter(combined_csv_path, engine='openpyxl')

# Loop through each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        file_path = os.path.join(directory, filename)

        # Check if the file is empty
        if os.path.getsize(file_path) == 0:
            print(f"Skipping empty file: {filename}")
            continue

        # Read each CSV file
        try:
            df = pd.read_csv(file_path)
            # Check if the DataFrame is empty
            if df.empty:
                print(f"No data in file: {filename}")
                continue

            # Add a column for the ticker based on the filename
            ticker = filename.split('.')[0]

            # Write DataFrame to a new sheet in the Excel file
            df.to_excel(writer, sheet_name=ticker, index=False)
            print(f"Data for {ticker} added to combined Excel file.")

        except pd.errors.EmptyDataError:
            print(f"Empty data error for file: {filename}")
            continue

# Save the combined Excel file
writer.close()
print(f"Combined data saved to {combined_csv_path}")
```

```python
import matplotlib.pyplot as plt

# Load the data from the NVDA sheet
file_path = '/content/drive/Mydrive/combined.xlsx'
nvda_data = pd.read_excel(file_path, sheet_name='NVDA')

# Convert the date column to datetime format
nvda_data['Date'] = pd.to_datetime(nvda_data['Date'])

# Filter the data for the specified date range
date_range = (nvda_data['Date'] >= '2024-05-20') & (nvda_data['Date'] <= '2024-05-24')
filtered_data = nvda_data[date_range]

# Calculate the moving averages
filtered_data['MA30'] = filtered_data['Close'].rolling(window=30).mean()
filtered_data['MA60'] = filtered_data['Close'].rolling(window=60).mean()

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(filtered_data['Date'], filtered_data['Close'], label='Close Price', marker='o')
plt.plot(filtered_data['Date'], filtered_data['MA30'], label='MA30', linestyle='--', color='orange')
plt.plot(filtered_data['Date'], filtered_data['MA60'], label='MA60', linestyle='--', color='green')

plt.title('NVDA Close Price with MA30 and MA60')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 올바른 파일 경로를 지정합니다.
file_path = '/content/drive/MyDrive/US MARKET/combined.xlsx'

# NVDA 시트 데이터를 로드합니다.
nvda_data = pd.read_excel(file_path, sheet_name='NVDA')

# Datetime 컬럼이 있는지 확인하고, Datetime 형식으로 변환합니다.
nvda_data['Datetime'] = pd.to_datetime(nvda_data['Datetime'])

# 특정 기간의 데이터를 필터링합니다.
start_date = '2024-05-20'
end_date = '2024-05-24'
date_range = (nvda_data['Datetime'] >= start_date) & (nvda_data['Datetime'] <= end_date)
filtered_data = nvda_data[date_range]

# 이동평균 계산
filtered_data['MA30'] = filtered_data['Close'].rolling(window=30).mean()
filtered_data['MA60'] = filtered_data['Close'].rolling(window=60).mean()

# 데이터 확인 (필터링된 데이터와 이동평균 컬럼이 추가된 것을 확인합니다)
print(filtered_data.head())

# 그래프를 그립니다.
plt.figure(figsize=(12, 6))
plt.plot(filtered_data['Datetime'], filtered_data['Close'], label='Close Price', marker='o')
plt.plot(filtered_data['Datetime'], filtered_data['MA30'], label='MA30', linestyle='--', color='orange')
plt.plot(filtered_data['Datetime'], filtered_data['MA60'], label='MA60', linestyle='--', color='green')

plt.title('NVDA Close Price with MA30 and MA60')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# 그래프 표시
plt.show()
```

```python
# 2024년 5월 23일까지의 데이터 필터링
end_date = '2024-05-23'
filtered_data = nvda_data[nvda_data['Datetime'].dt.date <= pd.to_datetime(end_date).date()]

# Reset index to ensure proper rolling calculation
filtered_data = filtered_data.reset_index(drop=True)

# Calculate the moving averages (30 minutes and 60 minutes)
filtered_data['MA30'] = filtered_data['Close'].rolling(window=30, min_periods=1).mean()
filtered_data['MA60'] = filtered_data['Close'].rolling(window=60, min_periods=1).mean()

# Plot the data for the specified date range with 30-min and 60-min moving averages
plt.figure(figsize=(12, 6))
plt.plot(filtered_data['Datetime'], filtered_data['Close'], label='Close Price', marker='o')
plt.plot(filtered_data['Datetime'], filtered_data['MA30'], label='MA30 (30 minutes)', linestyle='--', color='orange')
plt.plot(filtered_data['Datetime'], filtered_data['MA60'], label='MA60 (60 minutes)', linestyle='--', color='green')

plt.title('NVDA Close Price up to 2024-05-23 with MA30 and MA60')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
